```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
R_C = 2.0
R_I = 4.0

# Definio da funo
def f(x, R_C, R_I):
    if 0 < x <= R_C:
        return x**2
    elif R_C < x <= R_I:
        return (x - R_I)**2
    else:
        return np.nan

# Vetor de pontos
x_vals = np.linspace(0, R_I + 1, 500)
y_vals = [f(x, R_C, R_I) for x in x_vals]

# Plot
plt.figure(figsize=(7, 5))
plt.plot(x_vals, y_vals, label='$f(x)$', color='blue', linewidth=2)

# Marcao dos pontos de transio
plt.axvline(R_C, color='orange', linestyle='--', label='$R_C$')
plt.axvline(R_I, color='red', linestyle='--', label='$R_I$')

plt.title('Funo por partes $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
R_C = 2.0
R_I = 4.0

# Definio da funo
def f(x, R_C, R_I):
    if 0 < x <= R_C:
        return 10 * np.log10(x)
    elif R_C < x <= R_I:
        return 10 * np.log10(R_I - x)
    else:
        return np.nan

# Vetor de pontos
x_vals = np.linspace(0, R_I + 1, 500)
y_vals = [f(x, R_C, R_I) for x in x_vals]

# Plot
plt.figure(figsize=(7, 5))
plt.plot(x_vals, y_vals, label='$f(x)$', color='blue', linewidth=2)

# Marcao dos pontos de transio
plt.axvline(R_C, color='orange', linestyle='--', label='$R_C$')
plt.axvline(R_I, color='red', linestyle='--', label='$R_I$')

plt.title('Funo por partes $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
R_C = 2.0
R_I = 4.0
delta = 0.5   # amplitude da perturbao uniforme
N = 2

# Definio da funo
def f(x, R_C, R_I):
    noise = np.random.uniform(-delta, delta)
    if 0 < x <= R_C:
        return 10 * N * np.log10(x) + noise
    elif R_C < x <= R_I:
        return 10 * N * np.log10(R_I - x) + noise
    else:
        return np.nan

# Vetor de pontos
x_vals = np.linspace(0, R_I + 1, 500)
y_vals = [f(x, R_C, R_I) for x in x_vals]

# Plot
plt.figure(figsize=(7, 5))
plt.plot(x_vals, y_vals, label='$f(x)$', color='blue', linewidth=2)

# Marcao dos pontos de transio
plt.axvline(R_C, color='orange', linestyle='--', label='$R_C$')
plt.axvline(R_I, color='red', linestyle='--', label='$R_I$')

plt.title('Funo por partes $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
