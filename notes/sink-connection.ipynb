# Nota sobre conectividade ao sink

Nas simulações, observou-se que certas configurações de motes próximos ao *sink* podem degradar a performance da rede. Em particular, a presença de motes que:
- permanecem dentro de região de interferência;
- mantém conectividade direta e indireta com o sink;
tende a aumentar colisões, atrasos e perdas de pacote, reduzindo a eficiência global do sistema.

Essas observações motivaram uma análise geométrica, para determinar o número máximo de motes que podem coexistir em condição ótima ao redor do *sink*, respeitando simultaneamente as restrições de alcance de comunicação e de interferência. Isto é, investigando a relação entre o raio de comunicação $R_{\text{com}}$ e o raio de interferência $R_{\text{interf}}$.

>### *Definição*
>Sejam $0<R_{C}\le R_I$ e um ponto $\sigma\in\mathbb R^2$. 
>Dado um ponto $x\in\mathbb R^2$ denotamos por $(R_C,R_I)_x$ a faixa entre os circulos centrados em $x$ de raios $R_C$ e $R_I$ respectivamente.
>O subconjunto $P\subset\mathbb R^2$ de pontos que satisfaçam as seguintes condições:
>1. $\|p-\sigma\|\le R_C$, $\forall p\in P$.
>2. Para quaisquer $p\ne q\in P$, tem-se $p,q\notin\bigcup_{x\in P\cup\{\sigma\}}(R_C,R_I)_x$ e, também $\|p-q\|>R_C$.
>
>será chamado de *conjunto sem interferência*.

Asim queremos saber o número máximo $|P|$ de pontos sem interferência.

>### *Teorema*
>Dadas as hipóteses da definição acima. Seja $\rho=R_I/R_C$. Então o número máximo de pontos sem interferência é:
>$$
>n_{\max}(\rho)=\max\Big\{n\in\mathbb N, n\ge 1\;|\;2\sin\Big(\frac{\pi}{n}\Big)\ge\rho\Big\}
>$$
>com a convenção de que, se o conjunto entre chaves é vazio, então $n_{\max}=1$.

#### Demonstração

A prova tem três passos: (i) empurrar os pontos à borda, (ii) reduzir o problema a separação angular na circunferência, (iii) obter a cota superior por comprimento de arco.

**Passo 1: Redução à circunferência $\|p-\sigma\|=R_C$**

Se $p\in P$ satisfaz $\|p-\sigma\|<R_C$, considere o ponto $p'$ obtido por projetar $p$ radialmente sobre a circunferência:
$$
p' = \sigma + R_C\frac{p-\sigma}{\|p-\sigma\|}.
$$

Vamos mostrar que, para qualquer $q\in P\setminus\{p\}$, temos simultaneamente:
1. $\|p'-q\|\ge \|p-q\|$;
2. $p'\notin (R_C,R_I)_q$.

*(1) Monotonicidade da distância ao longo do raio*

Considere o triângulo formado por $\sigma,p,q$. Seja $\theta$ o ângulo em $\sigma$ entre os vetores $p-\sigma$ e $q-\sigma$. Pela lei dos cossenos:
$$
\|p-q\|^2
  = \|p-\sigma\|^2 + \|q-\sigma\|^2 - 2\|p-\sigma\|\|q-\sigma\|\cos\theta.
$$
Fixando $q$ e $\theta$, essa expressão é uma função quadrática crescente em $\|p-\sigma\|$ sempre que $\cos\theta \le 1$.  
Logo, ao substituir $\|p-\sigma\|$ por $R_C > \|p-\sigma\|$, temos:
$$
\|p'-q\| \ge \|p-q\|.
$$
Portanto, a projeção radial não diminui a distância de $p$ a nenhum outro ponto $q$.

*(2) Preservação da não interferência*

Como cada região de interferência $(R_C,R_I)_x$ é delimitada por círculos concêntricos em torno de $x$, e $\|p'-x\|\ge \|p-x\|$ para todo $x\ne p$, segue que, se $p$ estava fora de $(R_C,R_I)_x$, então $p'$ também estará fora, pois a distância ao centro $x$ aumentou ou permaneceu igual.  
Assim, $p'$ não entra em nenhuma faixa proibida $(R_C,R_I)_x$ de pontos de $P\cup\{\sigma\}$.

Assim, concluímos que podemos substituir $p$ por $p'$ sem violar nenhuma das condições de não interferência.  
Repetindo o procedimento para todos os pontos de $P$ com $\|p-\sigma\|<R_C$, obtemos um novo conjunto $P'\subseteq\{x\in\mathbb R^2:\|x-\sigma\|=R_C\}$, com $|P'|=|P|$, que também é um conjunto sem interferência.

**Passo 2: Redução à separação angular na circunferência**

Conforme verificamos no passo 1, podemos supor sem perda de generalidade que todos os pontos de $P$ estão sobre a circunferência
$$
\|p-\sigma\| = R_C, \quad \forall\,p\in P.
$$

Sejam $p,q\in P$ dois pontos distintos, e seja $\theta(p,q)\in[0,\pi]$ o ângulo central entre eles, isto é, o ângulo no vértice $\sigma$ formado pelos vetores $p-\sigma$ e $q-\sigma$.

Pela lei dos cossenos, temos:
$$
\|p-q\|^2
  = 2R_C^2(1 - \cos\theta(p,q))
  = 4R_C^2 \sin^2\!\Big(\frac{\theta(p,q)}{2}\Big),
$$
de onde segue que
$$
\|p-q\| = 2R_C \sin\!\Big(\frac{\theta(p,q)}{2}\Big).
$$

A restrição de não interferência impõe que
$$
\|p-q\| \ge R_I,
$$
o que equivale a
$$
2R_C \sin\!\Big(\frac{\theta(p,q)}{2}\Big) \ge R_I
\quad\Longleftrightarrow\quad
\theta(p,q) \ge 2\arcsin\!\Big(\frac{R_I}{2R_C}\Big)
= 2\arcsin\!\Big(\frac{\rho}{2}\Big).
$$

Portanto, quaisquer dois pontos distintos de $P$ devem estar separados, ao redor de $\sigma$, por um ângulo central mínimo
$$
\alpha(\rho) := 2\arcsin\!\Big(\frac{\rho}{2}\Big).
$$

**Passo 3: Cota superior por separação angular**

Ordenemos os pontos de $P$ por seus ângulos polares em torno de $\sigma$:
$$
0 \le \theta_1 < \theta_2 < \dots < \theta_{|P|} < 2\pi.
$$

Cada arco entre dois pontos consecutivos tem comprimento angular pelo menos $\alpha(\rho)$, de modo que a soma total dos arcos não ultrapassa $2\pi$:
$$
|P|\cdot \alpha(\rho) \le 2\pi.
$$

Daí obtemos a cota superior:
$$
|P| \le \frac{2\pi}{\alpha(\rho)}
= \frac{2\pi}{2\arcsin(\rho/2)}
= \frac{\pi}{\arcsin(\rho/2)}.
$$

Tomando a parte inteira inferior, obtemos:
$$
|P| \le \Big\lfloor \frac{2\pi}{2\arcsin(\rho/2)} \Big\rfloor.
$$

**Passo 4: Realizabilidade (existência de configuração ótima)**

A cota superior é atingida quando os pontos de $P$ estão uniformemente distribuídos sobre a circunferência, isto é, formam um poligono regular (com $n$ vértices) centrado em $\sigma$ com ângulo central entre vizinhos igual a:
$$
\frac{2\pi}{n}.
$$

Nesse arranjo, a menor distância entre vértices consecutivos é:
$$
d_{\min}(n) = 2R_C \sin\!\Big(\frac{\pi}{n}\Big).
$$

Logo, a condição de não interferência será satisfeita se, e somente se,
$$
d_{\min}(n) \ge R_I
\quad\Longleftrightarrow\quad
2R_C \sin\!\Big(\frac{\pi}{n}\Big) \ge R_I.
$$

Substituindo $\rho = R_I / R_C$, temos:
$$
2\sin\!\Big(\frac{\pi}{n}\Big) \ge \rho.
$$

O maior inteiro $n$ que satisfaz essa desigualdade é exatamente:
$$
n_{\max}(\rho)
= \max\Big\{n\in\mathbb N, n\ge 1\;|\;2\sin\Big(\frac{\pi}{n}\Big)\ge\rho\Big\}.
$$

Esse valor é atingido por um $n$-gono regular centrado em $\sigma$, provando que a cota é ótima.

Combinando os passos anteriores, obtemos que:

- Todos os pontos podem ser projetados sobre a circunferência de raio $R_C$ sem violar as restrições;
- As restrições de não interferência equivalem a uma separação angular mínima $\alpha(\rho)$ entre pontos;
- A separação angular limita o número máximo de pontos à condição $n\cdot \alpha(\rho)\le 2\pi$;
- E a configuração regular atinge essa cota.

Portanto,
$$
\boxed{
n_{\max}(\rho)
= \max\Big\{n\in\mathbb N,\ n\ge 1\;|\;
2\sin\Big(\frac{\pi}{n}\Big)\ge\rho\Big\}.
}
$$

---

> **Observação:**  
> Os pontos de quebra para $n=6,5,4,3,2,1$ ocorrem respectivamente em
> $$
> \rho \in \{1,\ 2\sin(\pi/5),\ \sqrt{2},\ \sqrt{3},\ 2\},
> $$
> reproduzindo as situações geométricas observadas nas simulações:
> hexágono regular (sem interferência no limite), pentágono, quadrado, triângulo, dois pontos diametralmente opostos e, por fim, apenas um ponto possível quando $\rho>2$.



## Visualização Geométrica

```python
import math
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge

def n_max_points(rho: float, n_upper: int = 40) -> int:
    """Maior n tal que 2*sin(pi/n) >= rho (ou 1 se nenhum n≥2 satisfaz)."""
    for n in range(n_upper, 0, -1):
        if 2 * math.sin(math.pi / n) >= rho:
            return n
    return 1

def polygon_on_circle(n: int, Rc: float = 1.0, angle0: float = 0.0):
    """Coordenadas de um n-gono regular sobre a circunferência de raio Rc."""
    if n == 1:
        return [(Rc, 0.0)]
    return [
        (Rc * math.cos(angle0 + 2 * math.pi * k / n),
         Rc * math.sin(angle0 + 2 * math.pi * k / n))
        for k in range(n)
    ]

def chord_min(n: int, Rc: float = 1.0) -> float:
    """Distância mínima entre vértices vizinhos no n-gono regular."""
    if n <= 1:
        return 0.0
    return 2 * Rc * math.sin(math.pi / n)

def plot_case(rho: float, Rc: float = 1.0):
    """
    Desenha:
      • regiões proibidas (anel entre Rc e Ri) em laranja translúcido
      • círculos Rc (azul) e Ri (vermelho)
      • conexões entre o centro e os vértices
    """
    n = n_max_points(rho)
    print(f'n={n}')
    Ri = rho * Rc
    pts = polygon_on_circle(n, Rc)
    dmin = chord_min(n, Rc)

    fig, ax = plt.subplots()
    fig.set_size_inches(6, 6)
    ax.set_aspect("equal")
    ax.set_facecolor("white")

    # conjunto (centro + vértices)
    all_points = [(0.0, 0.0)] + pts

    # === regiões proibidas ===
    for (x, y) in all_points:
        forbidden = Wedge(center=(x, y), r=Ri, theta1=0, theta2=360,
                          width=Ri - Rc, facecolor="orange", alpha=0.25, edgecolor=None)
        ax.add_patch(forbidden)

    # === círculos Rc e Ri em todos os pontos ===
    for (x, y) in all_points:
        circ_rc = Circle((x, y), Rc, color="tab:blue", fill=False, lw=1.5, ls="--", alpha=0.6)
        circ_ri = Circle((x, y), Ri, color="tab:red", fill=False, lw=1.0, ls="--", alpha=0.6)
        ax.add_patch(circ_rc)
        ax.add_patch(circ_ri)

    # === conexões e pontos ===
    for (x, y) in pts:
        ax.plot([0, x], [0, y], color="tab:blue", lw=1)
        ax.plot(x, y, "o", color="tab:blue", markersize=5)

    ax.plot(0, 0, "o", color="black", markersize=6)

    # limites
    M = 1.25 * max(Ri, Rc)
    ax.set_xlim(-M, M)
    ax.set_ylim(-M, M)
    ax.set_xticks([])
    ax.set_yticks([])

    # título e legenda
    ax.set_title(
        f"rho = R_i/R_c = {rho:.3f} | n_max = {n} | d_min = {dmin:.3f} | Rc={Rc:.3f} | Ri={Ri:.3f}",
        fontsize=10,
        color="black"
    )

    ax.legend(
        handles=[
            plt.Line2D([], [], color="tab:blue", lw=1.5, ls="--", label="R_c"),
            plt.Line2D([], [], color="tab:red", lw=1.0, ls="--", label="R_i"),
            plt.Line2D([], [], color="orange", lw=6, alpha=0.3, label="Região proibida")
        ],
        loc="upper right", fontsize=8
    )

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    Rc = 1.0
    cases = [
        0.999,
        1.15,
        1.30,
        1.60,
        1.80,
        2.10,
    ]

    for rho in cases:
        plot_case(rho=rho, Rc=Rc)
```

```python
import math
from pathlib import Path
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge

# =====================================================================================
# Cores (tema azul)
# =====================================================================================
COLOR_VERTICES = "#1f77b4"
COLOR_EDGES = "#1f77b4"
COLOR_CENTER = "#000000"
COLOR_RC = "#1f77b4"
COLOR_RI = "#d62728"
COLOR_FORBIDDEN = "#ff7f0e"

# =====================================================================================
# Funções auxiliares
# =====================================================================================
def n_max_points(rho: float, n_upper: int = 60) -> int:
    for n in range(n_upper, 1, -1):
        if 2 * math.sin(math.pi / n) >= rho:
            return n
    return 1

def polygon_on_circle(n: int, Rc: float = 1.0, angle0: float = 0.0):
    if n == 1:
        return [(Rc, 0.0)]
    return [
        (Rc * math.cos(angle0 + 2 * math.pi * k / n),
         Rc * math.sin(angle0 + 2 * math.pi * k / n))
        for k in range(n)
    ]

def chord_min(n: int, Rc: float = 1.0) -> float:
    if n <= 1:
        return 0.0
    return 2 * Rc * math.sin(math.pi / n)

def fig_to_frame(fig):
    fig.canvas.draw()
    w, h = fig.canvas.get_width_height()
    buf = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    return buf.reshape(h, w, 3)

def align_polygons(V_old, V_new):
    m = len(V_old)
    k = len(V_new)
    if m != k:
        return V_new[:]
    Vn = np.array(V_new)
    Vo = np.array(V_old)
    best_shift = 0
    best_score = float("inf")
    for s in range(k):
        rolled = np.roll(Vn, shift=s, axis=0)
        score = np.linalg.norm(rolled - Vo)
        if score < best_score:
            best_score = score
            best_shift = s
    return list(np.roll(Vn, shift=best_shift, axis=0))

def morph_vertices(V_old, V_new, steps=6):
    m = len(V_old)
    k = len(V_new)
    Vo = np.array(V_old, dtype=float)
    Vn = np.array(V_new, dtype=float)
    if m == k:
        Vn = np.array(align_polygons(V_old, V_new))

    L = max(m, k)
    def pad_vertices(V, target_len):
        if len(V) == target_len:
            return V
        idx = np.linspace(0, len(V)-1, num=target_len, endpoint=True)
        idx_floor = np.floor(idx).astype(int)
        return V[idx_floor]

    Vo_pad = pad_vertices(Vo, L)
    Vn_pad = pad_vertices(Vn, L)
    alpha_old = np.ones(L, dtype=float) if m >= L else np.concatenate([np.ones(m), np.zeros(L-m)])
    alpha_new = np.ones(L, dtype=float) if k >= L else np.concatenate([np.ones(k), np.zeros(L-k)])

    frames = []
    for s in range(steps):
        t = s / (steps - 1) if steps > 1 else 1.0
        verts = (1 - t) * Vo_pad + t * Vn_pad
        alpha = (1 - t) * alpha_old + t * alpha_new
        frames.append({"verts": [tuple(p) for p in verts], "alpha": alpha.tolist()})
    return frames

# =====================================================================================
# Plot simplificado (sem arestas do polígono)
# =====================================================================================
def plot_frame(rho: float, Rc: float, verts, alpha, dpi: int = 120,
               xlim=None, ylim=None):
    Ri = rho * Rc
    n = max(1, len(verts))
    dmin = chord_min(n, Rc)

    fig, ax = plt.subplots(figsize=(6, 6), dpi=dpi)
    ax.set_aspect("equal")
    ax.set_facecolor("white")

    all_points = [(0.0, 0.0)] + verts

    for (x, y) in all_points:
        if Ri > Rc:
            forbidden = Wedge(center=(x, y), r=Ri, theta1=0, theta2=360,
                              width=Ri - Rc, facecolor=COLOR_FORBIDDEN, alpha=0.25, edgecolor=None)
            ax.add_patch(forbidden)

    for (x, y) in all_points:
        circ_rc = Circle((x, y), Rc, fill=False, lw=1.5, ls="--", alpha=0.7, edgecolor=COLOR_RC)
        circ_ri = Circle((x, y), Ri, fill=False, lw=1.0, ls="--", alpha=0.7, edgecolor=COLOR_RI)
        ax.add_patch(circ_rc)
        ax.add_patch(circ_ri)

    # Somente arestas do centro aos vértices
    for (a, (x, y)) in zip(alpha, verts):
        ax.plot([0, x], [0, y], lw=1.0, color=COLOR_EDGES, alpha=0.8 * a)
        ax.plot(x, y, "o", color=COLOR_VERTICES, markersize=4 + 2*a, alpha=a)

    ax.plot(0, 0, "o", color=COLOR_CENTER, markersize=5)

    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(f"ρ = {rho:.3f} | n = {n} | d_min = {dmin:.3f}", fontsize=10)

    frame = fig_to_frame(fig)
    plt.close(fig)
    return frame

# =====================================================================================
# Geração do GIF
# =====================================================================================
Rc = 1.0
rho_start = 0.999
rho_end = 2.10
base_frames = 180
fps = 24
smooth_steps = 8
dpi_render = 140
Ri_max = rho_end * Rc
Mfix = 1.25 * max(Ri_max, Rc)
XLIM = (-Mfix, Mfix)
YLIM = (-Mfix, Mfix)

def rho_t(t, T): return rho_start + (rho_end - rho_start) * (t / (T - 1))

frames = []
rho0 = rho_t(0, base_frames)
n0 = n_max_points(rho0)
V_prev = polygon_on_circle(n0, Rc)
alpha_prev = [1.0] * len(V_prev)

for t in range(base_frames):
    rho = rho_t(t, base_frames)
    n_now = n_max_points(rho)
    V_now = polygon_on_circle(n_now, Rc)
    if len(V_now) != len(V_prev):
        morph = morph_vertices(V_prev, V_now, steps=smooth_steps)
        for step in morph:
            frame = plot_frame(rho, Rc, step["verts"], step["alpha"], dpi=dpi_render, xlim=XLIM, ylim=YLIM)
            frames.append(frame)
    else:
        frame = plot_frame(rho, Rc, V_now, [1.0]*len(V_now), dpi=dpi_render, xlim=XLIM, ylim=YLIM)
        frames.append(frame)
    V_prev = V_now
    alpha_prev = [1.0] * len(V_prev)

out_path = Path("./rho_evolution_center_edges.gif")
imageio.mimsave(out_path.as_posix(), frames, duration=1.0/fps, loop=0)

out_path.as_posix()
```
