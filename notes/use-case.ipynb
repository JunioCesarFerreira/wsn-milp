```python
import math
import itertools
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import networkx as nx

# ===================== Dados da cena (do seu exemplo) =====================
rects_m = [[-96.31147540983606, -169.672131147541, 63.114754098360656, 56.55737704918033], [-178.27868852459017, -169.672131147541, 31.147540983606557, 58.196721311475414], [102.04918032786885, -168.0327868852459, 62.295081967213115, 58.196721311475414], [-17.62295081967213, -168.0327868852459, 35.24590163934426, 57.37704918032787], [-139.75409836065575, -168.0327868852459, 31.9672131147541, 28.688524590163937], [25.81967213114754, -167.21311475409837, 65.57377049180329, 58.196721311475414], [-138.11475409836066, -130.32786885245903, 30.327868852459016, 30.327868852459016], [52.868852459016395, -100.0, 21.311475409836067, 57.37704918032787], [-96.31147540983606, -100.0, 62.295081967213115, 56.55737704918033], [106.14754098360656, -98.36065573770492, 58.196721311475414, 57.37704918032787], [-179.0983606557377, -98.36065573770492, 67.21311475409836, 23.770491803278688], [-20.901639344262296, -97.54098360655738, 38.52459016393443, 59.83606557377049], [-11.885245901639344, -39.34426229508197, 24.59016393442623, 41.80327868852459], [-97.95081967213115, -30.327868852459016, 63.9344262295082, 56.55737704918033], [-178.27868852459017, -30.327868852459016, 68.0327868852459, 59.83606557377049], [107.78688524590164, -29.508196721311478, 59.016393442622956, 57.37704918032787], [35.65573770491803, -26.229508196721312, 59.016393442622956, 57.37704918032787], [-178.27868852459017, 27.868852459016395, 68.0327868852459, 47.540983606557376], [107.78688524590164, 29.508196721311478, 60.65573770491803, 49.18032786885246], [-99.59016393442623, 30.327868852459016, 66.39344262295081, 50.81967213114754], [32.37704918032787, 32.786885245901644, 64.75409836065575, 52.459016393442624], [111.88524590163935, 67.21311475409836, 59.016393442622956, 26.229508196721312], [36.47540983606557, 91.80327868852459, 63.9344262295082, 52.459016393442624], [-100.40983606557377, 127.04918032786885, 68.85245901639344, 88.52459016393443], [-178.27868852459017, 127.04918032786885, 68.0327868852459, 90.16393442622952], [-20.901639344262296, 131.9672131147541, 42.622950819672134, 92.62295081967213], [111.88524590163935, 135.24590163934425, 31.9672131147541, 57.37704918032787], [74.18032786885246, 135.24590163934425, 22.950819672131146, 31.147540983606557]]
motes_m = [(30.737704918032787, -231.14754098360658), (143.85245901639345, -229.5081967213115), (-88.11475409836066, -228.68852459016395), (-22.54098360655738, -214.75409836065575), (98.77049180327869, -200.81967213114754), (-130.7377049180328, -200.0), (-183.19672131147541, -197.95081967213116), (133.60655737704917, -163.11475409836066), (-58.606557377049185, -162.29508196721312), (-141.39344262295083, -149.18032786885246), (22.54098360655738, -146.72131147540983), (-99.59016393442623, -135.24590163934425), (80.73770491803279, -125.0), (-130.7377049180328, -94.67213114754098), (133.60655737704917, -92.62295081967213), (-60.24590163934426, -90.16393442622952), (32.37704918032787, -68.0327868852459), (174.18032786885246, -55.73770491803279), (-181.96721311475412, -39.34426229508197), (-29.098360655737707, -36.0655737704918), (-154.5081967213115, -25.0), (-86.47540983606558, -23.770491803278688), (73.77049180327869, -22.950819672131146), (0.0, -0.0), (104.09836065573771, 15.573770491803279), (172.13114754098362, 16.80327868852459), (-133.19672131147541, 32.786885245901644), (-47.950819672131146, 33.60655737704918), (45.49180327868853, 37.704918032786885), (-106.9672131147541, 58.196721311475414), (150.0, 85.24590163934427), (-186.0655737704918, 92.21311475409836), (-26.639344262295083, 95.49180327868852), (88.9344262295082, 97.1311475409836), (27.459016393442624, 122.54098360655738), (-147.54098360655738, 130.7377049180328), (130.32786885245903, 138.9344262295082)]
SINK = (0.0, 0.0)
RC = 100.0
XLIM = (-201.0655737704918, 189.18032786885246)
YLIM = (-246.14754098360658, 153.9344262295082)

# ===================== Geometria bsica =====================
def rect_to_bounds(r):
    # retngulo no seu formato (x0, y_top, w, h)
    x0, y_top, w, h = r
    xmin = x0
    xmax = x0 + w
    ymax = y_top
    ymin = y_top - h
    return xmin, ymin, xmax, ymax

def point_in_rect(x, y, r):
    xmin, ymin, xmax, ymax = rect_to_bounds(r)
    return (xmin <= x <= xmax) and (ymin <= y <= ymax)

def seg_intersects_rect(p, q, r):
    # checa interseo segmento vs retngulo (inclusive borda)
    xmin, ymin, xmax, ymax = rect_to_bounds(r)
    # recorte trivial: ambos dentro -> tratar como "bloqueado"
    if point_in_rect(p[0], p[1], r) or point_in_rect(q[0], q[1], r):
        return True
    # CohenSutherland rpido + checagem com 4 arestas
    rx = [(xmin, ymin), (xmax, ymin), (xmax, ymax), (xmin, ymax)]
    edges = list(zip(rx, rx[1:] + rx[:1]))
    for a, b in edges:
        if segments_intersect(p, q, a, b):
            return True
    return False

def ccw(a, b, c):
    return (c[1]-a[1])*(b[0]-a[0]) > (b[1]-a[1])*(c[0]-a[0])

def segments_intersect(a, b, c, d):
    # interseo geral (inclui colinearidade na borda)
    def on_seg(p, q, r):
        return (min(p[0], r[0]) - 1e-9 <= q[0] <= max(p[0], r[0]) + 1e-9 and
                min(p[1], r[1]) - 1e-9 <= q[1] <= max(p[1], r[1]) + 1e-9)
    # orientao
    def orient(p, q, r):
        val = (q[1]-p[1])*(r[0]-q[0]) - (q[0]-p[0])*(r[1]-q[1])
        return 0 if abs(val) < 1e-12 else (1 if val > 0 else 2)
    o1 = orient(a, b, c); o2 = orient(a, b, d)
    o3 = orient(c, d, a); o4 = orient(c, d, b)
    if (o1 != o2) and (o3 != o4):
        return True
    # colinear e sobreposto
    if o1 == 0 and on_seg(a, c, b): return True
    if o2 == 0 and on_seg(a, d, b): return True
    if o3 == 0 and on_seg(c, a, d): return True
    if o4 == 0 and on_seg(c, b, d): return True
    return False

def los_blocked(p, q, rects):
    return any(seg_intersects_rect(p, q, r) for r in rects)

# ===================== Grafo de conectividade =====================
def build_graph(motes, rc, rects, block_by_rects=True):
    G = nx.Graph()
    # adiciona SINK como n 0, depois motes como 1..N
    nodes = [SINK] + list(motes)
    for idx, pos in enumerate(nodes):
        G.add_node(idx, pos=pos)
    for i, j in itertools.combinations(range(len(nodes)), 2):
        xi, yi = nodes[i]; xj, yj = nodes[j]
        if (xi - xj)**2 + (yi - yj)**2 <= rc*rc:
            if block_by_rects and los_blocked((xi, yi), (xj, yj), rects):
                continue
            G.add_edge(i, j, dist=math.hypot(xi-xj, yi-yj))
    return G

# ===================== Mtricas de conectividade =====================
def connectivity_report(G):
    n = G.number_of_nodes()
    comps = list(nx.connected_components(G))
    comp_sizes = sorted([len(c) for c in comps], reverse=True)
    sink_comp = next((c for c in comps if 0 in c), set([0]))
    sink_size = len(sink_comp)
    frac_sink = sink_size / n

    # hops e dimetro no componente do sink
    sub = G.subgraph(sink_comp)
    sp = dict(nx.single_source_shortest_path_length(sub, 0))
    hops_stats = {
        "min": min(sp.values()) if sp else 0,
        "max": max(sp.values()) if sp else 0,
        "avg": sum(sp.values())/len(sp) if sp else 0.0
    }
    diameter = nx.diameter(sub) if sub.number_of_nodes() >= 2 else 0

    # pontos de corte e pontes (no grafo completo)
    arts = list(nx.articulation_points(G))
    bridges = list(nx.bridges(G))

    degs = dict(G.degree())
    avg_deg = sum(degs.values())/n if n else 0.0

    return {
        "num_nodes": n,
        "num_edges": G.number_of_edges(),
        "num_components": len(comps),
        "component_sizes_desc": comp_sizes,
        "sink_component_size": sink_size,
        "fraction_connected_to_sink": frac_sink,
        "avg_degree": avg_deg,
        "hops_stats_in_sink_component": hops_stats,
        "diameter_sink_component": diameter,
        "articulation_points": arts,
        "bridges": bridges,
    }

# ===================== Cobertura por amostragem =====================
def coverage_estimate(motes, rc, rects, xlim, ylim, step=5.0, require_los=False):
    xs = frange(xlim[0], xlim[1], step)
    ys = frange(ylim[0], ylim[1], step)
    total_free = 0
    covered = 0
    covered_points = []  # para plot opcional
    for x in xs:
        for y in ys:
            # ignora pontos dentro de obstculos
            if any(point_in_rect(x, y, r) for r in rects):
                continue
            total_free += 1
            # checa cobertura por algum mote
            for (mx, my) in motes:
                if (mx - x)**2 + (my - y)**2 <= rc*rc:
                    if require_los and los_blocked((mx, my), (x, y), rects):
                        continue
                    covered += 1
                    covered_points.append((x, y))
                    break
    cov = covered / total_free if total_free else 0.0
    return cov, covered_points

def frange(a, b, step):
    x = a
    # garante incluso do limite superior com tolerncia
    while x <= b + 1e-9:
        yield x
        x += step

# ===================== Visualizao =====================
def plot_scene(show_edges=True, show_coverage=False, coverage_pts=None, figsize=(6.5, 6.5)):
    fig, ax = plt.subplots(figsize=figsize); ax.set_aspect('equal')

    # obstculos
    for (x0, y_top, w, h) in rects_m:
        ax.add_patch(Rectangle((x0, y_top - h), w, h, facecolor='0.6', edgecolor='0.6'))

    # cobertura
    if show_coverage and coverage_pts:
        cx, cy = zip(*coverage_pts)
        ax.scatter(cx, cy, s=5, alpha=0.25, marker='s', linewidths=0)

    # arestas
    if show_edges:
        # reconstroi grafo (com bloqueio por LOS para desenhar fiel ao reporte)
        G = build_graph(motes_m, RC, rects_m, block_by_rects=True)
        for u, v in G.edges():
            xi, yi = G.nodes[u]['pos']; xj, yj = G.nodes[v]['pos']
            ax.plot([xi, xj], [yi, yj], '-', linewidth=1.0, color='red', alpha=0.8, zorder=2)

    # motes (pretos)
    xs = [p[0] for p in motes_m]; ys = [p[1] for p in motes_m]
    ax.scatter(xs, ys, s=25, c='k', zorder=3)

    # sink (quadrado vazio)
    size = 16.0
    ax.scatter([SINK[0]],[SINK[1]], s=30, c='k', zorder=4)
    ax.add_patch(Rectangle((SINK[0]-size/2, SINK[1]-size/2), size, size, fill=False, linewidth=1.8))

    ax.set_xlim(*XLIM); ax.set_ylim(*YLIM)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# ===================== Execuo =====================
if __name__ == "__main__":
    # 1) Grafo com bloqueio por obstculos
    G_block = build_graph(motes_m, RC, rects_m, block_by_rects=True)
    report_block = connectivity_report(G_block)
    print("== Conectividade (com bloqueio por obstculos / LOS) ==")
    for k, v in report_block.items():
        print(f"{k}: {v}")

    # 2) Grafo ignorando obstculos (canal livre)
    G_free = build_graph(motes_m, RC, rects_m, block_by_rects=False)
    report_free = connectivity_report(G_free)
    print("\n== Conectividade (ignorando obstculos) ==")
    for k, v in report_free.items():
        print(f"{k}: {v}")

    # 3) Cobertura por amostragem
    #    (ajuste step para mais fino/grosso; require_los=True exige LOS at o ponto)
    step = 4.0
    cov_free, pts_free = coverage_estimate(motes_m, RC, rects_m, XLIM, YLIM, step=step, require_los=False)
    cov_los,  pts_los  = coverage_estimate(motes_m, RC, rects_m, XLIM, YLIM, step=step, require_los=True)

    print(f"\nCobertura estimada (ignorando obstculos na recepo): {cov_free*100:.2f}%")
    print(f"Cobertura estimada (exigindo LOS at um mote):        {cov_los*100:.2f}%")

    # 4) Plotar cena + cobertura (troque entre pts_free/pts_los)
    plot_scene(show_edges=True, show_coverage=True, coverage_pts=pts_los, figsize=(7.2, 7.2))
```
