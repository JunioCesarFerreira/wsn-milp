```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle

# =========================================================
# Configuraes
# =========================================================
R_com = 25
R_inter = 40
SHOW_RADII = True
SHOW_SOJOURN_RADII = False

# =========================================================
# Gerador de clusters sintticos
# =========================================================
def make_cluster(center, n=10, spread=8):
    return np.random.normal(loc=center, scale=spread, size=(n, 2))

# clusters de sensores
cluster1 = make_cluster(center=[20, 20], n=12, spread=4)
cluster2 = make_cluster(center=[80, 20], n=10, spread=5)
cluster3 = make_cluster(center=[50, 70], n=14, spread=6)

sensors = np.vstack([cluster1, cluster2, cluster3])

# sojourn locations
sojourns = np.array([
    [15, 15],
    [30, 12],
    [25, 25],
    [70, 15],
    [80, 30],
    [70, 55],
    [50, 65],
    [30, 60],
    [50, 80],
])

# base do sink
base_pos = np.array([10, 80])

# todos os ns da trajetria (0 = base, 1..n = sojourns)
nodes = np.vstack([base_pos, sojourns])
n_nodes = nodes.shape[0]

# =========================================================
# Construo de um grafo mais "complexo"
# =========================================================
edges = []

# 1) Arestas em cadeia para garantir conectividade
for i in range(n_nodes - 1):
    edges.append((i, i + 1))

# 2) Arestas extras para dar ramificaes
#    (ligamos cada n ao prximo e ao "prximo do prximo" quando possvel)
for i in range(n_nodes):
    j = i + 2
    if j < n_nodes:
        edges.append((i, j))

# 3) Arestas da base (n 0) para alguns sojourns mais distantes (efeito estrela)
for j in [3, 5, 7]:
    if j < n_nodes:
        edges.append((0, j))

# Remover duplicadas (i,j) e (j,i) mantidos como um s
edges_unique = set()
for i, j in edges:
    if i < j:
        edges_unique.add((i, j))
    else:
        edges_unique.add((j, i))
edges = list(edges_unique)

# =========================================================
# Plot
# =========================================================
fig, ax = plt.subplots(figsize=(6.5, 6))

# sensores
ax.scatter(
    sensors[:, 0], sensors[:, 1],
    s=45, c="black", label="Sensores fixos"
)

# sojourns
ax.scatter(
    sojourns[:, 0], sojourns[:, 1],
    s=80, edgecolor="purple", facecolor="lavender",
    marker="P", label="Sojourn locations"
)

# base do sink
ax.scatter(
    base_pos[0], base_pos[1],
    s=200, c="red", marker="s", label="Base / sink"
)
ax.add_patch(Rectangle(
    (base_pos[0] - 3, base_pos[1] - 3),
    6, 6,
    linewidth=1.2, edgecolor="darkred", facecolor="none"
))

# arestas tracejadas conectando base + sojourns (grafo mais complexo)
for i, j in edges:
    x1, y1 = nodes[i]
    x2, y2 = nodes[j]
    ax.plot(
        [x1, x2], [y1, y2],
        linestyle="--", linewidth=1.2, color="purple"
    )

# =========================================================
# Raios (opcionais)
# =========================================================
if SHOW_RADII:
    for x, y in sensors:
        ax.add_patch(Circle((x, y), R_com, color="blue",  alpha=0.03))
        ax.add_patch(Circle((x, y), R_inter, color="orange", alpha=0.02))

if SHOW_SOJOURN_RADII:
    for x, y in sojourns:
        ax.add_patch(Circle((x, y), R_com, color="blue", alpha=0.05))

# =========================================================
# Esttica
# =========================================================
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_aspect("equal", adjustable="box")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Sink mvel: base, sojourn locations e trajetria")

ax.legend(
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    borderaxespad=0.,
    fontsize=8
)

plt.tight_layout()
plt.show()
```
